```python
from evds import evdsAPI
import pandas as pd
import numpy as np
evds = evdsAPI('kvlUQV03Zp')
```

```python
term = '2023-06'
```

```python
term_year = term[:4]
term_month = term[-2:]
term_end_day = pd.Period('{}-01-{}'.format(term_month, term_year), freq='M').end_time.day

start_date_str = '01-{}-{}'.format(term_month, term_year)
end_date_str = '{}-{}-{}'.format(term_end_day, term_month, term_year)
```

```python
data_list = ['TP.DK.AUD.A.YTL', 'TP.DK.CAD.A.YTL', 'TP.DK.CHF.A.YTL', 'TP.DK.DKK.A.YTL', 'TP.DK.EUR.A.YTL', 'TP.DK.GBP.A.YTL', 'TP.DK.JPY.A.YTL', 'TP.DK.KWD.A.YTL', 'TP.DK.NOK.A.YTL', 'TP.DK.SAR.A.YTL', 'TP.DK.SEK.A.YTL', 'TP.DK.USD.A.YTL']
df_evds = evds.get_data(data_list, startdate = start_date_str, enddate = end_date_str, aggregation_types='avg', frequency = 5)
df_evds
```

```python
df = pd.DataFrame()
df['Döviz'] = data_list
df['Döviz'] = df['Döviz'].apply(lambda x: x[6:9])
df[term] = list(df_evds.iloc[0, 1:])
df = df.sort_values('Döviz').set_index('Döviz')
df
```

```python
df['Fix'] = (np.round(df[term] * 1_000_000, -1)).astype(int)
df
```
